```python
from lxml import etree
import requests
import pandas as pd
import re
# 设置最大爬取页数
max_page = 20

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36'
}
```

```python
all_title = []
all_time = []
for page in range(1, max_page + 1):

    url = f'https://guba.eastmoney.com/list,zsgjdjia,f_{page}.html'
    data = requests.get(url,timeout = 10, headers = headers)
    # 这一行设置非常重要 如果不加 encoding 的话可能会出现乱码
    data.encoding = 'utf-8'
    
    data_e = etree.HTML(data.text)

    title = data_e.xpath("//tr[contains(@class,'listitem')]//div[@class = 'title']//a//text()")
    time = data_e.xpath("//tr[contains(@class,'listitem')]//div[@class = 'update']//text()")
    #read = data_e.xpath("//tr[contains(@class,'listitem')]//div[@class = 'read']//text()")
    all_title += title
    all_time += time
    #print(len(all_time),len(all_title))
        

print(all_title)
```
